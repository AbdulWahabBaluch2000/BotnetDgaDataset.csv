```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
df = pd.read_csv("BotnetDgaDataset.csv")  # Replace with actual file path
```

```python
# Display shape of dataset
print("Shape of Dataset:", df.shape)

# Display dataset info
print("\nDataset Info:")
print(df.info())

# Display statistical summary
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values Per Feature:")
print(df.isnull().sum())

# Check for infinite values
print("\nInfinite Values Per Feature:")
print((df == np.inf).sum())

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Verify missing values are handled
print("\nMissing Values After Handling:")
print(df.isnull().sum().sum())  # Should print 0 if all are handled
```

```python
# Check for duplicate rows
print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Verify duplicate removal
print("\nDuplicate Rows After Removal:", df.duplicated().sum())
```

```python
# Compute correlation matrix
corr_matrix = df.corr().abs()

# Identify highly correlated features
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]

# Drop correlated features
df_selected = df.drop(columns=to_drop)
print(f"\nRemoved {len(to_drop)} highly correlated features.")

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_selected.corr(), cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Check if 'Label' or 'Class' column exists
class_col = [col for col in df_selected.columns if 'class' in col.lower() or 'label' in col.lower()]

if class_col:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df_selected[class_col[0]])
    plt.title("Class Distribution")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("\nNo class label found in dataset.")
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_selected)
plt.xticks(rotation=90)
plt.title("Boxplot Before Handling Outliers")
plt.show()
```

```python
df_capped = df_selected.copy()
for col in df_capped.select_dtypes(include=["number"]).columns:
    upper_limit = df_capped[col].quantile(0.99)
    df_capped[col] = np.where(df_capped[col] > upper_limit, upper_limit, df_capped[col])

plt.figure(figsize=(15, 6))
sns.boxplot(data=df_capped)
plt.xticks(rotation=90)
plt.title("Boxplot After Capping Outliers (99th Percentile)")
plt.show()
```

```python
df_log = df_capped.copy()
df_log[df_log.columns] = np.log1p(df_log[df_log.columns])

plt.figure(figsize=(15, 6))
sns.boxplot(data=df_log)
plt.xticks(rotation=90)
plt.title("Boxplot After Log Transformation")
plt.show()
```

```python
# Check if label column exists before applying Random Forest
if class_col:
    X = df_selected.drop(columns=class_col[0])
    y = df_selected[class_col[0]]

    model = RandomForestClassifier()
    model.fit(X, y)

    feature_importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": model.feature_importances_
    }).sort_values(by="Importance", ascending=False)

    # Plot feature importance
    plt.figure(figsize=(10, 5))
    sns.barplot(x="Importance", y="Feature", data=feature_importance.head(20))
    plt.title("Top 20 Feature Importance (Random Forest)")
    plt.show()
else:
    print("\nSkipping Feature Importance Evaluation - No Class Label Found.")
```

```python
# Standardize features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)

# Apply PCA (retain 95% variance)
pca = PCA(n_components=0.95)
df_pca = pca.fit_transform(df_scaled)

print(f"\nOriginal features: {df_selected.shape[1]}, Reduced features: {df_pca.shape[1]}")
```
