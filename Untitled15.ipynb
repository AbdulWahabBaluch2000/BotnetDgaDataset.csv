```python
dataset_path = '/content/drive/MyDrive/SOCOFing/Real'

image_files = sorted([f for f in os.listdir(dataset_path) if f.endswith('.BMP')])[:5]
print("Sample files:", image_files)
```

```python
import cv2

# Create ORB detector
orb = cv2.ORB_create()

# Function to extract keypoints and descriptors from an image
def extract_features(img_path):
    img = cv2.imread(os.path.join(dataset_path, img_path), cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (200, 200))  # Resize for consistency
    keypoints, descriptors = orb.detectAndCompute(img, None)
    return keypoints, descriptors
```

```python
# Test feature extraction on the first fingerprint
test_image = image_files[0]
kp, desc = extract_features(test_image)

print(f"Image: {test_image}")
print(f"Number of keypoints: {len(kp)}")
print(f"Descriptor shape: {desc.shape if desc is not None else 'None'}")
```

```python
# Brute-force matcher for Hamming distance
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

def match_fingerprints(desc1, desc2):
    if desc1 is None or desc2 is None:
        return 0
    matches = bf.match(desc1, desc2)
    return len(matches)

# Collect scores for genuine and impostor comparisons
true_matches = []
false_matches = []

for i in range(len(image_files)):
    kp1, desc1 = extract_features(image_files[i])
    for j in range(len(image_files)):
        kp2, desc2 = extract_features(image_files[j])
        score = match_fingerprints(desc1, desc2)
        if i == j:
            true_matches.append(score)
        else:
            false_matches.append(score)
```

```python
# Set a threshold for matching score
threshold = 30  # Tune this if needed

# Calculate FNMR (False Non-Match Rate)
FNMR = sum([s < threshold for s in true_matches]) / len(true_matches)

# Calculate FMR (False Match Rate)
FMR = sum([s >= threshold for s in false_matches]) / len(false_matches)

print(f"False Non-Match Rate (FNMR): {FNMR:.2f}")
print(f"False Match Rate (FMR): {FMR:.2f}")
```

```python
import matplotlib.pyplot as plt  # ðŸ”§ Re-import here to fix the error
from sklearn.metrics import roc_curve
import numpy as np

# Combine true labels and match scores
y_true = [1]*len(true_matches) + [0]*len(false_matches)
y_scores = true_matches + false_matches

# Calculate ROC metrics
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
fnr = 1 - tpr

# Calculate Equal Error Rate (EER)
eer_index = np.nanargmin(np.abs(fnr - fpr))
eer = fpr[eer_index]
eer_threshold = thresholds[eer_index]

print(f"Equal Error Rate (EER): {eer:.4f} at threshold = {eer_threshold:.2f}")

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label='ROC Curve', color='blue')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.scatter(fpr[eer_index], tpr[eer_index], color='red', label=f'EER = {eer:.2f}')
plt.xlabel('False Positive Rate (FMR)')
plt.ylabel('True Positive Rate (1 - FNMR)')
plt.title('ROC Curve for Fingerprint Matching')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define altered dataset path
altered_path = '/content/drive/MyDrive/SOCOFing/Altered/Altered-Easy/'

# Load altered image files (sample subset)
altered_files = sorted([f for f in os.listdir(altered_path) if f.endswith('.BMP')])[:20]  # sample

# Match altered fingerprints to their real counterparts
spoof_successes = 0
total_attempts = 0

for altered_img in altered_files:
    # Try to find matching real image by name
    real_img = altered_img.replace('_Altered.Easy', '')

    if real_img in image_files:
        # Extract features
        _, desc_real = extract_features(real_img)
        _, desc_altered = extract_features(os.path.join(altered_path, altered_img))

        # Compute match score
        score = match_fingerprints(desc_real, desc_altered)

        # Spoof success if match score is above threshold
        if score >= threshold:
            spoof_successes += 1
        total_attempts += 1

# Calculate ATSR
ATSR = spoof_successes / total_attempts if total_attempts > 0 else 0
print(f"Ability to Spoof Rate (ATSR): {ATSR:.2f}")
```

```python
# Simulate FTER (Failure to Enrol Rate)
enrol_failures = 0
total_enrol_attempts = len(image_files)

for img_name in image_files:
    _, desc = extract_features(img_name)
    if desc is None or len(desc) == 0:
        enrol_failures += 1

FTER = enrol_failures / total_enrol_attempts
print(f"Failure to Enrol Rate (FTER): {FTER:.2f}")
```

```python
# Simulate ATV (Ability to Verify Rate)
verification_attempts = 0
successful_verifications = 0

for i in range(len(image_files)):
    _, desc1 = extract_features(image_files[i])
    for j in range(len(image_files)):
        _, desc2 = extract_features(image_files[j])
        verification_attempts += 1
        if desc1 is not None and desc2 is not None:
            successful_verifications += 1

ATV = successful_verifications / verification_attempts
print(f"Ability to Verify Rate (ATV): {ATV:.2f}")
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
from google.colab import drive
drive.mount('/content/drive')

dataset_path = '/content/drive/MyDrive/SOCOFing/Real'

# Load images and labels
images = []
labels = []

for file in sorted(os.listdir(dataset_path))[:1000]:  # limit for speed
    if file.endswith('.BMP'):
        img = cv2.imread(os.path.join(dataset_path, file), cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (100, 100)) / 255.0  # normalize
        images.append(img)
        label = file.split('__')[0]  # get subject ID
        labels.append(label)

images = np.array(images).reshape(-1, 100, 100, 1)  # CNN-ready shape
labels = np.array(labels)
print(f"Loaded {len(images)} images.")
```

```python
# Encode person ID labels to integers
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)
num_classes = len(np.unique(encoded_labels))
y = to_categorical(encoded_labels, num_classes)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(images, y, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Flatten(input_shape=(100, 100, 1)),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
# Plot accuracy
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Final accuracy
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Predict on test set
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = np.argmax(y_test, axis=1)

# Confusion matrix
cm = confusion_matrix(y_true_labels, y_pred_labels)

# Plot the confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Fingerprint Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification report

print(classification_report(y_true_labels, y_pred_labels))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(100, 100, 1)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')  # softmax for multi-class
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('CNN Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
